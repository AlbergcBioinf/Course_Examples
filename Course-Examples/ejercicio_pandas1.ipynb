# 7. Crear una función que reciba una estación, un contaminante (MAGNITUD) y un rango de fechas y devuelva una serie con las emisiones del contaminante dado en la estación y rango de fechas dado.

```python
def f(estacion, contaminante, inicio, fin):
    mascara = (emisiones["MAGNITUD"]==contaminante) & (emisiones["ESTACION"]==estacion) & (emisiones["FECHA"]>=inicio) & (emisiones["FECHA"]<=fin) 
    serie =  emisiones[mascara].sort_values("FECHA")[["VALOR", "FECHA"]]
    serie = serie.set_index("FECHA")
    return serie["VALOR"]
```

```python
inicio = dt.datetime(2016,1,1)
fin = dt.datetime(2016,1,31)

f(4, 1, inicio, fin )
```
